```python
"""
Predator–Prey Dynamic System (Lotka–Volterra Model)
---------------------------------------------------
This script models and visualizes population dynamics between two interacting species:
prey (e.g., rabbits) and predators (e.g., foxes).

It uses a system of ordinary differential equations (ODEs) to simulate how
both populations evolve over time based on biological interaction rates.

Author: [Your Name]
Date: [Month, Year]
"""

# Import required libraries
import numpy as np                      # For numerical operations
from scipy.integrate import odeint      # For solving differential equations
import matplotlib.pyplot as plt         # For plotting the results


# -------------------------------
# Define the dynamic system model
# -------------------------------
def predator_prey(state, t, alpha, beta, delta, gamma):
    """
    Defines the Lotka–Volterra predator–prey model equations.

    Parameters:
        state : list or array
            Current populations [x, y] where:
            x = prey population
            y = predator population
        t : float
            Time variable (required for odeint)
        alpha : float
            Natural growth rate of prey (when no predators)
        beta : float
            Rate at which predators consume prey
        delta : float
            Rate at which predators reproduce per prey eaten
        gamma : float
            Natural death rate of predators

    Returns:
        list : [dx/dt, dy/dt] representing population change rates
    """
    x, y = state
    dxdt = alpha * x - beta * x * y        # Change in prey population
    dydt = delta * x * y - gamma * y       # Change in predator population
    return [dxdt, dydt]


# -------------------------------
# Model Parameters
# -------------------------------
alpha = 0.6   # Prey birth rate
beta = 0.02   # Predation rate
delta = 0.01  # Predator reproduction rate
gamma = 0.4   # Predator death rate


# -------------------------------
# Initial Conditions
# -------------------------------
x0 = 40   # Initial prey population
y0 = 9    # Initial predator population
initial_state = [x0, y0]


# -------------------------------
# Time Range for Simulation
# -------------------------------
t = np.linspace(0, 200, 1000)  # Simulate from 0 to 200 time units with 1000 points


# -------------------------------
# Solve the System of Equations
# -------------------------------
solution = odeint(predator_prey, initial_state, t, args=(alpha, beta, delta, gamma))
x, y = solution.T  # Transpose results for plotting


# -------------------------------
# Plot Population Changes Over Time
# -------------------------------
plt.figure(figsize=(10, 6))
plt.plot(t, x, label="Prey Population", linewidth=2)
plt.plot(t, y, label="Predator Population", color='orange', linewidth=2)
plt.title("Dynamic System in Biology: Predator–Prey Model", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# -------------------------------
# Plot Phase-Space (Predator vs Prey)
# -------------------------------
plt.figure(figsize=(7, 6))
plt.plot(x, y, color='purple', linewidth=2)
plt.title("Phase-Space Plot: Predator vs Prey Dynamics", fontsize=14)
plt.xlabel("Prey Population (x)", fontsize=12)
plt.ylabel("Predator Population (y)", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
